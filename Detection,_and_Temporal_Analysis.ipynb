1) What is Dimensionality Reduction? Why is it important in machine
learning?

Dimensionality Reduction is the process of reducing the number of input variables (features) in a dataset while retaining as much important information as possible.
Example-Suppose you have a dataset with 100 features (columns). Many of them might be:
Highly correlated (carry similar information)
Irrelevant or noisy
Dimensionality reduction techniques (like PCA – Principal Component Analysis) can transform these 100 features into, say, 10 new features that still capture most of the variance in the data.

Dimensionality Reduction Importance-
i)Improves Model Performance
ii)Reduces Overfitting
iii)Speeds Up Computation
iv)Simplifies Visualization


2)  Name and briefly describe three common dimensionality reduction
techniques.

1. Principal Component Analysis (PCA)
Type: Linear technique
How it works: PCA transforms the original correlated features into a smaller set of uncorrelated variables called principal components, which capture the maximum variance (information) in the data.
Use case: Widely used for data compression, visualization and noise reduction.

2. Linear Discriminant Analysis (LDA)
Type: Supervised linear technique
How it works: LDA finds new feature axes that maximize the separation between different classes while minimizing variation within each class.
Use case: Mainly used in classification problems like face recognition or text classification.

3. t-Distributed Stochastic Neighbor Embedding (t-SNE)
Type: Non-linear technique
How it works: t-SNE reduces dimensions by converting high-dimensional distances into probabilities and trying to preserve the local structure of the data in a low-dimensional space.
Use case: Excellent for visualizing high-dimensional data (e.g., word embeddings, image data) in 2D or 3D.

3) What is clustering in unsupervised learning? Mention three popular
clustering algorithms.

Clustering is a technique in unsupervised learning used to group similar data points together based on their features — without using predefined labels.

Three Popular Clustering Algorithms:

i)K-Means Clustering
-Divides data into K clusters by minimizing the distance between points and their cluster’s centroid.
-Fast and efficient, but you must specify the number of clusters (K) beforehand.

ii)Hierarchical Clustering
-Builds a tree-like structure (dendrogram) of clusters by successively merging or splitting them based on similarity.
-No need to predefine K, and it helps visualize relationships between clusters.

iii)DBSCAN (Density-Based Spatial Clustering of Applications with Noise)
-Groups points that are closely packed together and marks outliers as noise.
-Can detect clusters of arbitrary shape and handles noise/outliers well.

4)Explain the concept of anomaly detection and its significance

Anomaly Detection is the process of identifying data points, events, or patterns that deviate significantly from the normal behavior of a dataset.

Significance of Anomaly Detection

-Improves Security:
Detects fraud, intrusions, and system failures early.

-Enhances Decision-Making:
Identifies unusual trends or risks that need immediate attention.

-Maintains Quality Control:
Detects defective items or errors in manufacturing or data processing.

-Prevents Loss:
Early detection of anomalies can save time, money, and resources.

5) List and briefly describe three types of anomaly detection techniques.

1)Statistical Methods
Concept: These techniques assume that normal data follows a specific statistical distribution (e.g., Gaussian/Normal distribution).
How it works: Data points that fall far from the expected range (for example, more than 3 standard deviations from the mean) are marked as anomalies.
Example: Z-score, Grubbs’ test, and Box plot methods.
Use case: Useful when data distribution is known and simple.

2. Machine Learning–Based Methods
Concept: These methods learn patterns of “normal” data and flag instances that deviate significantly.
How it works: Algorithms like Isolation Forest, One-Class SVM, or Autoencoders detect anomalies based on how different a data point is from learned patterns.
Use case: Effective for complex, high-dimensional data where statistical assumptions don’t hold.

3. Distance- or Density-Based Methods
Concept: These rely on the distance or density of data points.
How it works: Points that are far away from others (low density regions) are considered anomalies.
Examples:
K-Nearest Neighbors (KNN) – detects anomalies by measuring distance to neighbors.
DBSCAN – identifies low-density points as noise.
Use case: Works well when clusters or spatial structures exist in data.

6) What is time series analysis? Mention two key components of time series
data.

Time Series Analysis is a method of analyzing data points that are collected or recorded in chronological order (over time) to identify meaningful patterns, trends, and relationships.

Two Key Components of Time Series Data:

i)Trend
The long-term movement or direction in the data over time.
It shows whether values are increasing, decreasing, or remaining stable.
Example: A steady rise in housing prices over several years.

ii)Seasonality
The repeating short-term pattern or fluctuation that occurs at regular intervals (e.g., daily, monthly, yearly).
Example: Ice cream sales peaking every summer.

7)Describe the difference between seasonality and cyclic behavior in time
series.

Seasonality refers to regular and predictable fluctuations that occur at fixed time intervals, such as daily, monthly, or yearly. These patterns are often influenced by calendar-related factors like weather, holidays, or seasons—for example, higher ice cream sales every summer or increased electricity usage in winter.
In contrast, cyclic behavior represents irregular, long-term fluctuations that do not follow a fixed schedule. Cycles are often linked to broader economic or business conditions, such as periods of growth and recession in the economy. While seasonality is typically short-term and repetitive, cyclic behavior is longer-term and less predictable.

8)Write Python code to perform K-means clustering on a sample dataset.


```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Step 1: Create a sample dataset
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.6, random_state=0)

# Step 2: Visualize the data
plt.scatter(X[:, 0], X[:, 1], s=30)
plt.title("Sample Data")
plt.show()

# Step 3: Apply K-Means Clustering
kmeans = KMeans(n_clusters=4, random_state=0)
kmeans.fit(X)

# Step 4: Get cluster labels and centroids
y_kmeans = kmeans.predict(X)
centers = kmeans.cluster_centers_

# Step 5: Visualize the clusters
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=30, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')
plt.title("K-Means Clustering Result")
plt.show()
```

9) What is inheritance in OOP? Provide a simple example in Python.

Inheritance is a core concept in Object-Oriented Programming (OOP) that allows a class (called the child class or subclass) to inherit attributes and methods from another class (called the parent class or superclass).

It promotes code reusability, extensibility, and helps maintain a clear hierarchical structure between classes.

```python
# Parent class
class Animal:
    def speak(self):
        print("This animal makes a sound")

# Child class inherits from Animal
class Dog(Animal):
    def speak(self):
        print("The dog barks")

# Another child class
class Cat(Animal):
    def speak(self):
        print("The cat meows")

# Create objects
dog = Dog()
cat = Cat()

# Call methods
dog.speak()   # Output: The dog barks
cat.speak()   # Output: The cat meows
```

10) How can time series analysis be used for anomaly detection?


Time series analysis can be used for anomaly detection by identifying data points or patterns that deviate significantly from the normal temporal behavior of the series. Since time series data is ordered chronologically, anomalies are detected not just by their values but also by how they behave over time compared to historical trends, seasonality, and expected patterns.
Example:
-In a server monitoring system, time series analysis of CPU usage can reveal sudden spikes or drops that don’t match normal daily patterns — these would be anomalies.
-In finance, it can detect unusual stock price movements that deviate from expected trends.
