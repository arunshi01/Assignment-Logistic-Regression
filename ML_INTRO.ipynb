Question 1: Explain the differences between AI, ML, Deep Learning (DL), and Data
Science (DS).


1. Artificial Intelligence (AI)

Definition: AI is a broad field of computer science focused on creating systems that can perform tasks that typically require human intelligence.

Goal: To enable machines to think, reason, learn and make decisions.

Examples: Chatbots, self-driving cars, recommendation systems, speech recognition.

2. Machine Learning (ML)

Definition: A subset of AI where machines learn from data without being explicitly programmed.

Goal: To develop algorithms that improve their performance as they are exposed to more data.

Examples: Spam email filtering, credit card fraud detection, predictive maintenance.

Key Feature: Focuses on patterns and predictions from data.

3. Deep Learning (DL)

Definition: A specialized subset of ML that uses neural networks with multiple layers (deep neural networks) to model complex patterns.

Goal: To handle high-dimensional data like images, audio, and text with greater accuracy.

Examples: Face recognition, speech-to-text conversion, autonomous vehicle vision systems.

Key Feature: Requires large datasets and high computational power (e.g., GPUs).

4. Data Science (DS)

Definition: A multidisciplinary field that extracts insights from structured and unstructured data using statistics, ML, data visualization and domain expertise.

Goal: To analyze data and provide actionable insights for decision-making.

Examples: Business analytics dashboards, customer segmentation, A/B testing.

Scope: Includes data collection, cleaning, analysis, visualization, and modeling (often using AI/ML).

Question 2: What are the types of machine learning? Describe each with one
real-world example.

1. Supervised Learning
The model is trained on labeled data (input-output pairs). It learns the relationship between features (input) and the target variable (output).
Ex- Email Spam Detection – Model trained on emails labeled as spam or not spam to classify incoming emails.

2. Unsupervised Learning
The model is trained on unlabeled data and tries to find hidden patterns or groupings.
Example:
Customer Segmentation – Grouping customers into clusters (e.g., high spenders, occasional buyers) for targeted marketing.

3. Reinforcement Learning
An agent learns to make decisions by interacting with an environment and receiving rewards or penalties based on actions.
Eg- Self-Driving Cars – AI agent learns to navigate by receiving positive rewards for safe driving and penalties for collisions.

4. Semi-Supervised Learning (Hybrid)
Uses a small amount of labeled data and a large amount of unlabeled data to improve learning.
Eg- Medical Diagnosis – A few labeled X-rays (diseased vs. healthy) combined with a large dataset of unlabeled scans to improve disease detection.

Question 3: Define overfitting, underfitting, and the bias-variance tradeoff in machine learning.

**Overfitting**
Definition: When a model learns the training data too well, including noise and random fluctuations and performs poorly on unseen data.
Example: A decision tree that memorizes training samples instead of generalizing.

**Underfitting**
Definition: When a model is too simple to capture the underlying patterns in the data.
Example: Using linear regression for a highly nonlinear problem.

**Bias-Variance Tradeoff**
Bias: Error from overly simplistic assumptions in the model (leads to underfitting).


Question 4: What are outliers in a dataset, and list three common techniques for
handling them.

Outliers are data points that deviate significantly from the rest of the dataset.

Three Common Techniques to Handle Outliers
1. Removal (Exclusion)
When outliers are due to data entry errors or irrelevant extreme values.
Example:
Removing a value like 5000 kg in a dataset of human weights (clearly an error).

2. Transformation
Method:
Apply mathematical transformations to reduce the impact of extreme values (e.g., log, square root).
Example:
Using log transformation for income data to reduce the effect of extreme salaries.

3. Capping (Winsorization)
Method:
Replace extreme values with a certain percentile (e.g., 1st or 99th percentile).
Example:
Any value above the 99th percentile replaced with the value at the 99th percentile.

Question 5: Explain the process of handling missing values and mention one
imputation technique for numerical and one for categorical data.

Handling Missing Values in a Dataset
1. Identifying Missing Values
Use functions like isnull() in Python (Pandas) or profiling reports.
Check the percentage of missing data per column to decide treatment strategy.

2. Understanding Missingness
MCAR (Missing Completely at Random): No relationship to other data.
MAR (Missing at Random): Related to other observed variables.
MNAR (Missing Not at Random): Related to the value itself (e.g., income not reported by high earners).

3. Common Strategies
Deletion
Remove rows or columns with excessive missing data (e.g., > 50% missing).
Risk: Can lead to data loss.
Imputation
Filling missing values with estimated values based on other data.

**Imputation Techniques**
For Numerical Data
Mean/Median Imputation:
Replace missing values with the mean or median of the column.

Example:
If age values are missing, fill with median age (less sensitive to outliers).

For Categorical Data
Mode Imputation:
Replace missing values with the most frequent category.

Example:
If “Preferred Payment Method” is missing, fill with the most common value, e.g., "Credit Card".

8) Write a Python script to:
● Generate 1000 samples from a normal distribution.
● Introduce 50 random missing values.
● Fill missing values with the column mean.
● Plot a histogram before and after imputation.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Generate 1000 samples from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)  # mean=50, std=10

# Step 2: Introduce 50 random missing values
missing_indices = np.random.choice(1000, 50, replace=False)
data_with_nan = data.copy()
data_with_nan[missing_indices] = np.nan

# Step 3: Fill missing values with column mean
df = pd.DataFrame(data_with_nan, columns=['Values'])
mean_value = df['Values'].mean()
df_filled = df.fillna(mean_value)

# Step 4: Plot histogram before and after imputation
plt.figure(figsize=(10, 5))

# Before Imputation
plt.subplot(1, 2, 1)
plt.hist(df['Values'].dropna(), bins=30, edgecolor='black')
plt.title("Before Imputation")
plt.xlabel("Value")
plt.ylabel("Frequency")

# After Imputation
plt.subplot(1, 2, 2)
plt.hist(df_filled['Values'], bins=30, edgecolor='black')
plt.title("After Imputation")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```
